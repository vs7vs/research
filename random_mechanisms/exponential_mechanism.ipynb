```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exponential Mechanism for 2D points
def exponential_mechanism_2d(points, utility_function, epsilon, sensitivity):
    utilities = np.array([utility_function(point) for point in points])
    # Compute probabilities using the exponential mechanism formula
    probabilities = np.exp((epsilon * utilities) / (2 * sensitivity))
    probabilities /= np.sum(probabilities)  # Normalize probabilities
    return probabilities
```

```python
# Utility function (based on distance to a target point)
def utility_function(point, target=(5, 5)):
    return -np.linalg.norm(np.array(point) - np.array(target))  # Higher utility for closer points
```

```python
# Visualization of 2D points and their selection probabilities with a red cross at (5,5)
def visualize_exponential_mechanism_2d(points, probabilities):
    plt.figure(figsize=(8, 8))
    
    # Scatter plot of points with color indicating selection probabilities
    plt.scatter(points[:, 0], points[:, 1], c=probabilities, cmap='viridis', s=100, edgecolor='k')
    
    # Add a red cross at the target point (5, 5)
    plt.scatter(5, 5, color='red', marker='x', s=200, linewidth=3)
    
    # Add color bar and labels
    plt.colorbar(label='Selection Probability')
    plt.title('2D Points with Selection Probabilities (Exponential Mechanism)')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.grid(True)
    plt.legend()
    
    # Show plot
    plt.show()
```

```python
# Generate 2D uniformly distributed points
num_points = 200
points = np.random.uniform(0, 10, size=(num_points, 2))  # Uniformly distributed in 2D space [0, 10]x[0, 10]

# Parameters for the Exponential Mechanism
epsilon = 1.0  # Privacy parameter
sensitivity = 1.0  # Sensitivity of the utility function
target_point = (5, 5)  # Target point for utility calculation

# Apply the Exponential Mechanism
probabilities = exponential_mechanism_2d(points, lambda p: utility_function(p, target=target_point), epsilon, sensitivity)

# Visualize the points with their selection probabilities
visualize_exponential_mechanism_2d(points, probabilities)
```
